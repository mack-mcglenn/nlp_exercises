#### Put it all together

```python
url= 'https://codeup.com/blog/'
headers = {'User-Agent': 'Codeup Data Science'}
response= requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')

links = [link['href'] for link in soup.select('.more-link')]
articles =[]

for url in links:
    response= requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content)
    
    title = soup.find('h1', class_='entry-title').text
    content = soup.find('div', class_='entry-content').text.strip()
    content = content.replace(title, '')
    
    arts= {'title':title,
               'content': content}

    articles.append(arts)
   
    
```

```python
articles = pd.DataFrame(articles)
```

```python
cats = [li.text.lower() for li in soup.select('li')][1:]
cats[0]='national'

inshorts = []

for cat in cats:
    
    url= 'https://inshorts.com/en/read' + '/' + cat
    response = requests.get(url)
    soup= BeautifulSoup(response.content, 'html.parser')
    
    titles = [span.text for span in soup.find_all('span', itemprop='headline')[0]]
    content = [div.text for div in soup.find_all('div', itemprop='articleBody')[0]]
    
    for i in range(len(titles)):
        
        article = {
            'title': titles[i],
            'content': content,
            'category': cat
        }
        inshorts.append(article)
```

```python
inshorts[0:5]
```

```python
inshorts_df = pd.DataFrame(inshorts)
```
